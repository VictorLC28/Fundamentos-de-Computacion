```python
from automata.fa.nfa import NFA
from automata.fa.dfa import DFA
import graphviz

# Funcion para imprimir la informacion del automata y generar diagramas graficos
def imprimir_automata(afnd, afd, expresion_regular, palabras):
    print("\nExpresión Regular del lenguaje:", expresion_regular)

    # Imprimir la tupla que representa el AFND
    print("\nAFND:")
    print("Estados:", afnd.states)
    print("Símbolos de entrada (Alfabeto):", afnd.input_symbols)
    print("Transiciones:")
    for estado, transiciones in afnd.transitions.items():
        for simbolo, estados_siguientes in transiciones.items():
            print(f"  {estado} --{simbolo}--> {estados_siguientes}")
    print("Estado inicial:", afnd.initial_state)
    print("Estados finales:", afnd.final_states)

    # Imprimir la tupla que representa el AFD
    print("\nAFD:")
    print("Estados:", afd.states)
    print("Símbolos de entrada (Alfabeto):", afd.input_symbols)
    print("Transiciones:")
    for estado, transiciones in afd.transitions.items():
        for simbolo, estado_siguiente in transiciones.items():
            print(f"  {estado} --{simbolo}--> {estado_siguiente}")
    print("Estado inicial:", afd.initial_state)
    print("Estados finales:", afd.final_states)

    # Generar diagrama del AFD
    generar_diagrama_afd(afd, "AFD_diagrama")

    # Imprimir 8 palabras que pertenecen al lenguaje
    print("\n8 palabras que pertenecen al lenguaje:")
    for palabra in palabras:
        print(palabra)

# Funcion para generar diagrama grafico del AFD
def generar_diagrama_afd(afd, nombre_archivo):
    dot = graphviz.Digraph(format='png')
    dot.attr(rankdir='LR')  # Dirección del diagrama de izquierda a derecha

    # Agregar estados al grafico
    for estado in afd.states:
        estado = str(estado)  # Convertir a cadena
        if estado in afd.final_states:
            dot.node(estado, shape='doublecircle')  # Estados finales
        else:
            dot.node(estado, shape='circle')

    # Agregar transiciones
    for estado, transiciones in afd.transitions.items():
        for simbolo, estado_siguiente in transiciones.items():
            dot.edge(str(estado), str(estado_siguiente), label=simbolo)  # Convertir a cadena

    # Agregar estado inicial
    dot.node('start', shape='none', label='')
    dot.edge('start', str(afd.initial_state))

    # Guardar el archivo
    dot.render(filename=nombre_archivo, cleanup=True)
    print(f"El diagrama del AFD se ha guardado como '{nombre_archivo}.png'.")

# Función principal
def main():
    while True:
        print("\nSeleccione alguno de los siguientes lenguajes:")
        print("1. L(A)={bab,babab,baaabab,baaaabbbbab,baaaabbbbab, ...}")
        print("2. L(B)={101, 01011, 000001011111111, 00101, 0001011111, ...}")
        print("3. L(C)={aab, aaaaab, aabbbbbbb,aaaabbb,aabbbbbb, ...}")
        print("4. L(D)={1, 011,01101,0101011010101, 010101101010101,1010101, ...}")
        print("5. L(E)={aba,abb,abc,abca,abcccccb,abccccc,abcccca,abccccccccccca, ...}")
        print("6. Salir del programa")

        opcion = input("Ingrese el número de la opción deseada (1, 2, 3, 4, 5 o 6): ")

        if opcion == "1":
            expresion_regular = "ba* b* ab"
            afnd = NFA(
                states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6'},
                input_symbols={'a', 'b'},
                transitions={
                    'q0': {'b': {'q1'}},
                    'q1': {'a': {'q1', 'q2'}, 'b': {'q1'}},
                    'q2': {'a': {'q3'}, 'b': {'q2'}},
                    'q3': {'b': {'q4'}},
                    'q4': {'a': {'q5'}},
                    'q5': {'b': {'q6'}}
                },
                initial_state='q0',
                final_states={'q6'}
            )
            palabras = ["bab", "baab", "babab", "baabab", "baaaaabab", "baaaaabbab", "babab", "baaaabbbbbbbbbbab"]

        elif opcion == "2":
            expresion_regular = "0*1011*"
            afnd = NFA(
                states={'q0', 'q1', 'q2', 'q3', 'q4'},
                input_symbols={'0', '1'},
                transitions={
                    'q0': {'0': {'q0'}, '1': {'q1'}},
                    'q1': {'0': {'q2'}},
                    'q2': {'1': {'q3'}},
                    'q3': {'1': {'q4'}}
                },
                initial_state='q0',
                final_states={'q4'}
            )
            palabras = ["0101", "00101", "000101", "0000101", "1011", "1011111", "000000101111", "01011111"]

        elif opcion == "3":
            expresion_regular = "aa(a*+b*)b"
            afnd = NFA(
                states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8'},
                input_symbols={'a', 'b'},
                transitions={
                    'q0': {'a': {'q1'}},
                    'q1': {'a': {'q2'}},
                    'q2': {'a': {'q2', 'q3'}, 'b': {'q4'}},
                    'q3': {'b': {'q5'}},
                    'q4': {'a': {'q5'}},
                    'q5': {'b': {'q6'}}
                },
                initial_state='q0',
                final_states={'q6'}
            )
            palabras = ["aabb", "aaaaaab", "aaab", "aabbbbbbbbbb", "aab", "aabbbbbb", "aaaaaaab", "aaaab"]

        elif opcion == "4":
            expresion_regular = "(01)*(10)*1"
            afnd = NFA(
                states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5'},
                input_symbols={'0', '1'},
                transitions={
                    'q0': {'0': {'q1'}, '1': {'q3'}},
                    'q1': {'1': {'q2'}},
                    'q2': {'0': {'q1'}, '1': {'q3'}},
                    'q3': {'1': {'q4'}, '0': {'q5'}},
                    'q4': {'0': {'q5'}, '1': {'q4'}},
                    'q5': {'1': {'q4'}}
                },
                initial_state='q0',
                final_states={'q4'}
            )
            palabras = ["011", "0110101", "01011", "0101011", "01101", "0101101", "01101010101", "010101011"]

        elif opcion == "5":
            expresion_regular = "abc*(a + b + c)"
            afnd = NFA(
                states={'q0', 'q1', 'q2', 'q3', 'q4', 'q5'},
                input_symbols={'a', 'b', 'c'},
                transitions={
                    'q0': {'a': {'q1'}},
                    'q1': {'b': {'q2'}},
                    'q2': {'c': {'q2', 'q3'}},
                    'q3': {'a': {'q4'}, 'b': {'q4'}, 'c': {'q4'}}
                },
                initial_state='q0',
                final_states={'q4'}
            )
            palabras = ["aba", "abcb", "abcc", "abccc", "abca", "abccccccb", "abcca", "abcccb"]

        elif opcion == "6":
            print("Saliendo del programa. ¡Hasta luego!")
            break

        else:
            print("Opcion no valida")
            continue

        # Convertir el AFND a un AFD
        afd = DFA.from_nfa(afnd)

        # Imprimir el automata y la informacion
        imprimir_automata(afnd, afd, expresion_regular, palabras)

# Ejecutar el programa
main()
```
